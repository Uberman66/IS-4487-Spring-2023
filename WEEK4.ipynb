## Clean up the data
Did you notice anything unusual about the "house" amounts? 

How about the handset price and income?

```python
df1 = df[['customer_id', 'city', 'state_province','total_amount']].copy()

df1.head(15)
```

```python
df1['total_amount'] = df1['total_amount'].str.replace('$',' ')
df1['total_amount'] = pd.to_numeric(df1['total_amount'], errors='coerce')
df_clean = df1.dropna()
```

```python
df_clean.sort_values(by=['total_amount'])
```

# Add Derived Attribute

Add a column with the income in K (thousands)

```python
df_clean['income_k'] = (df_clean['income']/1000).round(2)
df_clean.head()
```
